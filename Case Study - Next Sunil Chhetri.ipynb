## Leveling the Playing Field - Making it equal

```python
fwd_mins = indian_forwards.groupby(["id", "name"])["minutes_played"].sum().reset_index()
fwd_mins.head(10)
```

```python
# simple plot of the minutes played
px.bar(x="name", y="minutes_played", text="minutes_played", data_frame=fwd_mins)
```

```python
# for analysis sakes lets keep only those players who have played in game
indian_forwards = indian_forwards[indian_forwards["minutes_played"] > 0]
```

```python
# per90 game stats logic
indian_forwards["goalsPer90"] = indian_forwards["events.goals"].divide(indian_forwards["minutes_played"]).multiply(90)
indian_forwards["assistsPer90"] = indian_forwards["events.assists"].divide(indian_forwards["minutes_played"]).multiply(90)
indian_forwards["shots_on_targetPer90"] = indian_forwards["events.shots_on_target"].divide(indian_forwards["minutes_played"]).multiply(90)
```

```python
# this list has columns of stuff we want to judge/assess as a per90 stat
per90Cols = ["events.goals", "events.assists", "events.shots", "events.shots_on_target", "events.chances_created",
             "events.fouls_suffered", "touches.total", 'touches.aerial_duel.won']

# this makes separate per90 columns with a suitable title and calculations
for col in per90Cols:
    indian_forwards[col + "Per90"] = indian_forwards[col].divide(indian_forwards["minutes_played"]).multiply(90)

indian_forwards
```

```python
# creating a list of per90 stats [similar to the column titles that were added]
cols_for_radar = []
for i in per90Cols:
    cols_for_radar.append(i + "Per90")
print(cols_for_radar)
```

```python
# this technique is called 'list comprehension' and does the same as above
cols_for_radar1 = [i + "Per90" for i in per90Cols]
print(cols_for_radar1)
```

```python
# Chhetri's ID is 19150 and these are his per90 stats
indian_forwards.loc[(indian_forwards["id"] == 19150), cols_for_radar].sum()
```

### Multiple Trace Radar Chart - to Perform a Comparative Analysis

```python
# this removes the duplicates of players [cos ISL 6 and ISL 7]
ind_fwds_id_names = indian_forwards_norm.drop_duplicates(subset=["id"])[["id", "name"]]
```

```python
ind_fwds_id_names.shape
```

```python
ind_fwds_id_names
```

```python
# cos normalization [you could put 1 but if you change normalization later it makes it dynamic]
isl_max = indian_forwards_norm[cols_for_radar].max().max()
```

```python
import plotly.graph_objects as go
```

```python
for i, row in ind_fwds_id_names.iterrows():
    if row["id"] == 19150:
        continue
    print(row["name"])
    # Initiate the plotly go figure
    fig = go.Figure()
    # Add Radar plots for different players:
    fig.add_trace(go.Scatterpolar(
        r=indian_forwards_norm.loc[(indian_forwards_norm["id"] == 19150), cols_for_radar].sum(),
        theta=cols_for_radar,
        fill='toself',
        name='Sunil Chhetri'))
    
    fig.add_trace(go.Scatterpolar(
        r=indian_forwards_norm.loc[(indian_forwards_norm["id"] == row["id"]), cols_for_radar].sum(),
        theta=cols_for_radar,
        fill='toself',
        name=row["name"]))
    
    # Additional properties for the plot:
    fig.update_layout(
        title="Sunil Chhetri vs " + row["name"],
        polar=dict(radialaxis=dict(visible=True, range=[0, isl_max])), showlegend=True)
    fig.show()
```
